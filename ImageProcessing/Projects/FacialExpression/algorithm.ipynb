```python
import cv2
import matplotlib.pyplot as plt
from deepface import DeepFace

# Read image
img = cv2.imread('dataset/face1.jpeg')

# Call imshow() using plt object
plt.imshow(img[:, :, ::-1])

# Display the image
plt.show()

# Analyze the image to detect emotions
result = DeepFace.analyze(img, actions=['emotion'])

# Get the emotion dictionary from the result
emotion_dict = result[0]['emotion']

# Get the emotion label with the highest percentage
emotion_label = max(emotion_dict.items(), key=lambda x: x[1])[0]

# Print emotion
print("Emotion:", {emotion_label})

# Draw the emotion label on the image
text = f"Emotion: {emotion_label}"
cv2.putText(img, text, (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)

# Display the image with the emotion label
cv2.imshow('Image with Emotion', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

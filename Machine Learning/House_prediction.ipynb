```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3, random_state = 42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
```

```python
reg.fit(X_train, y_train)
```

```python
reg.intercept_
```

```python
reg.coef_
```

```python
def calc(slope, intercept, hours):
    return slope * hours + intercept
```

```python
data = calc(reg.coef_, reg.intercept_,9.5)
```

```python
print(data)
```

```python
y_pred = reg.predict(X_test)
```

```python
print(y_pred)
```

```python
hpp = pd.DataFrame({'Actual': y_test.squeeze(),'Predicted': y_pred.squeeze()})
```

```python
hpp.plot.scatter(x = 'Actual', y = 'Predicted', title = 'PLOT BETWEEN AND PREDICTED')
```

```python
print(hpp)
```

```python
from sklearn.metrics import mean_absolute_error , mean_squared_error
```

```python
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
```

```python
print(f'Mean absolute error : {mae:.2f}')
print(f'Mean squared error : {mse:.2f}')
print(f'root Mean absolute error : {rmse:.2f}')
```

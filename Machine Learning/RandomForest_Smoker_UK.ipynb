```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
columns=['smoke','gender' ,'highest_qualification','nationality' ,'marital_status' ,'ethnicity' ,'region' ,'gross_income' ]
for col in columns:
  smoking[col]=le.fit_transform(smoking[col])
```

```python
smoking.head(3)
```

```python
from sklearn.preprocessing  import StandardScaler
```

```python
X=smoking.drop('smoke',axis=1)
Y=smoking['smoke']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,Y,random_state=42,test_size=0.33)
```

```python
se=StandardScaler()
X_train_scaled=se.fit_transform(x_train)
X_test_scaled=se.transform(x_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model=RandomForestClassifier()
model.fit(x_train,y_train)
pred_value=model.predict(x_train)
print("Accuaracy is",model.score(x_test,y_test)*100)
```

```python
sub=pd.DataFrame()
sub['smoker']=pred_value
```

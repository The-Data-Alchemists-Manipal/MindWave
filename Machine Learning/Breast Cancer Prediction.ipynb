```python
def convert(s):     #Converting chategorical data type to numerical data type
 # s = s.lower()
  if s == 'M':
    return 1
  elif s == 'B':
    return -1
```

```python
new_data = pd.read_csv('/content/drive/MyDrive/breast-cancer.csv',converters={'diagnosis':convert})
```

```python
sns.countplot(x=new_data['diagnosis'])
```

```python
new_data['diagnosis'].unique()
```

```python
X = new_data[['radius_mean','texture_mean','perimeter_mean','area_mean','smoothness_mean','compactness_mean','concavity_mean','concave points_mean','radius_worst','texture_worst','perimeter_worst','area_worst','smoothness_worst','compactness_worst','concavity_worst','concave points_worst','symmetry_worst','fractal_dimension_worst']]
y = new_data[['diagnosis']]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

**REGRESSION**


**Support Vector Regression**

```python
from sklearn.svm import SVC
classifier = SVC(kernel='linear')
classifier.fit(X_train, y_train)
```

```python
y_predictions = classifier.predict(X_test)
```

```python
y_predictions
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_predictions)
cm
```

```python
print("Accuracy:",accuracy_score(y_test, y_predictions))
```

```python
print(classification_report(y_test, y_predictions))
```

**Linear Regression**

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor = LinearRegression()
regressor.fit(X,y)
```

```python
y_pred = regressor.predict(X_test)
```

```python
y_pred
```

```python
df_pred = pd.DataFrame({'Actual':y_test.squeeze(), 'Predicted': y_pred.squeeze()})
print(df_pred)
```

**CLASSIFICATION**

**Naive Bayes Classifier**

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.naive_bayes import GaussianNB
clf = GaussianNB()
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
print("Accuracy:",accuracy_score(y_test, y_pred))
```

```python
print(classification_report(y_test, y_pred))
```

**Decision Tree Classifier**

```python
from sklearn.tree import export_graphviz
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
```

```python
model = DecisionTreeClassifier(criterion = "entropy", max_depth = 3, random_state = 42) # max_depth is the level of classification
```

```python
model.fit(X_train, y_train)
```

```python
model.predict(X_test)
```

```python
import graphviz
from sklearn import tree
#DOT data
dot_data = tree.export_graphviz(model, out_file = None, feature_names = ['radius_mean','texture_mean','perimeter_mean','area_mean','smoothness_mean','compactness_mean','concavity_mean','concave points_mean','radius_worst','texture_worst','perimeter_worst','area_worst','smoothness_worst','compactness_worst','concavity_worst','concave points_worst','symmetry_worst','fractal_dimension_worst'], 
                                class_names="diagnosis", filled=True)
```

```python
graph = graphviz.Source(dot_data, format = "png")
graph
```

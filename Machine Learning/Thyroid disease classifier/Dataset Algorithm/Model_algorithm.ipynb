```python
import numpy as np
import pandas as pd
import sklearn 
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv('thyroidDF.csv')
dataset.head()
```

```python
dataset.isna().sum()
```

```python
dataset.shape
```

```python
dataset=dataset.drop(['patient_id','referral_source'],axis='columns')
dataset.head()
```

```python
df=dataset.dropna(how='any',axis="rows",subset = ['sex'])
df.iloc[:,-2]=df.iloc[:,-2].fillna(0)
# df.iloc[:,[17,19,21,23]]=df.iloc[:,[17,19,21,23]].fillna()
```

```python
df.iloc[:,17]=df.iloc[:,17].fillna(np.mean(df.iloc[:,17]))
df.iloc[:,19]=df.iloc[:,19].fillna(np.mean(df.iloc[:,19]))
df.iloc[:,21]=df.iloc[:,21].fillna(np.mean(df.iloc[:,21]))
df.iloc[:,23]=df.iloc[:,23].fillna(np.mean(df.iloc[:,23]))
```

```python

df.iloc[:,-4]=df.iloc[:,-4].fillna(np.mean(df.iloc[:,-4]))
df.head()
```

```python
df.isna().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df.iloc[:,2]=le.fit_transform(df.iloc[:,2])
```

```python
df.iloc[:,3]=le.fit_transform(df.iloc[:,3])
df.iloc[:,4]=le.fit_transform(df.iloc[:,4])
df.iloc[:,5]=le.fit_transform(df.iloc[:,5])
df.iloc[:,6]=le.fit_transform(df.iloc[:,6])
df.iloc[:,7]=le.fit_transform(df.iloc[:,7])
df.iloc[:,8]=le.fit_transform(df.iloc[:,8])
df.iloc[:,9]=le.fit_transform(df.iloc[:,9])
```

```python
df.iloc[:,10]=le.fit_transform(df.iloc[:,10])
df.iloc[:,11]=le.fit_transform(df.iloc[:,11])
df.iloc[:,12]=le.fit_transform(df.iloc[:,12])
df.iloc[:,13]=le.fit_transform(df.iloc[:,13])
df.iloc[:,14]=le.fit_transform(df.iloc[:,14])
df.iloc[:,15]=le.fit_transform(df.iloc[:,15])
df.iloc[:,16]=le.fit_transform(df.iloc[:,16])
df.iloc[:,17]=le.fit_transform(df.iloc[:,17])
df.head()
```

```python
df.iloc[:,-3]=le.fit_transform(df.iloc[:,-3])
df.iloc[:,-5]=le.fit_transform(df.iloc[:,-5])
df.iloc[:,-7]=le.fit_transform(df.iloc[:,-7])
df.iloc[:,-9]=le.fit_transform(df.iloc[:,-9])
df.head()
```

```python
df.iloc[:,-11]=le.fit_transform(df.iloc[:,-11])
df.iloc[:,1]=le.fit_transform(df.iloc[:,1])

df.head()
```

```python
df.iloc[:,-1]=le.fit_transform(df.iloc[:,-1])

df.head()
```

```python
X=df.iloc[:,:-1]
Y=df.iloc[:,-1]
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=0)
```

```python
from sklearn.ensemble import RandomForestClassifier

model2=RandomForestClassifier(n_estimators=50)
model2.fit(X_train.astype('int'),Y_train.astype('int'))
```

```python
model2.score(X_test.astype('int'), Y_test.astype('int'))
```

```python
import joblib
joblib.dump(model2,"Trained_model.pkl")
```

#Cluster Analysis Using t-SNE

```python
from sklearn.manifold import TSNE


parameters = ['stick_pitch', 'IAS', 'AOA', 'Nz', 'roll', 'pitch']


tsne = TSNE(n_components=2, random_state=42)
reduced_data = tsne.fit_transform(df[parameters])


plt.scatter(reduced_data[:, 0], reduced_data[:, 1])
plt.xlabel('t-SNE Component 1')
plt.ylabel('t-SNE Component 2')
plt.title('Flight Test Data Clusters')
plt.show()
```

#Time Series Forecasting

```python
from statsmodels.tsa.arima.model import ARIMA


parameter = 'IAS'


train_data = df[parameter][:-50]
test_data = df[parameter][-50:]


model = ARIMA(train_data, order=(1, 0, 1))
model_fit = model.fit()


forecast_values = model_fit.forecast(steps=50)


plt.plot(train_data.index, train_data, label='Actual')
plt.plot(test_data.index, test_data, label='Actual (Test)')
plt.plot(test_data.index, forecast_values, label='Forecast')
plt.xlabel('Time')
plt.ylabel(parameter)
plt.title('Time-Series Forecasting')
plt.legend()
plt.show()
```

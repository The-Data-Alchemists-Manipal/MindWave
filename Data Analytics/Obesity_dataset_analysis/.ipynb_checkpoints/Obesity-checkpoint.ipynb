# About Data

This Analysis contains data for the estimation of obesity levels in people from the countries of Mexico, Peru and Colombia, with ages between 14 and 61 and diverse eating habits and physical condition, the information was processed obtaining 17 attributes and 2111 records, after a balancing process. 

The attributes related with eating habits are: 
Frequent consumption of high caloric food (FAVC), Frequency of consumption of vegetables (FCVC), Number of main meals (NCP), Consumption of food between meals (CAEC), Consumption of water daily (CH20), and Consumption of alcohol (CALC). 

The attributes related with the physical condition are: 
Calories consumption monitoring (SCC), Physical activity frequency (FAF), Time using technology devices (TUE), Transportation used (MTRANS), other variables obtained were: Gender, Age, Height and Weight. 

Finally, all data was labeled and the class variable NObesity was created with the values of: Insufficient Weight, Normal Weight, Overweight Level I, Overweight Level II, Obesity Type I, Obesity Type II and Obesity Type III, based on Equation (1) and information from WHO and Mexican Normativity. 
The data contains numerical data and continous data, so it can be used for analysis based on algorithms of classification, prediction, segmentation and association. Data is available in CSV format.

# Chi-Square Test 

The test is applied when you have two categorical variables from a single population. It is used to determine whether there is a significant association between the two variables

```python
import scipy.stats as stats
```

```python
dataset_table=pd.crosstab(dataset['Gender'],dataset['SMOKE'])
print(dataset_table)
     
```

```python
dataset_table.values
```

```python
Observed_Values = dataset_table.values 
print("Observed Values :-\n",Observed_Values)
```

```python
val=stats.chi2_contingency(dataset_table)
```

```python
val
```

```python
Expected_Values=val[3]
```

```python
Expected_Values
```

```python
no_of_rows=len(dataset_table.iloc[0:2,0])
no_of_columns=len(dataset_table.iloc[0,0:2])
ddof=(no_of_rows-1)*(no_of_columns-1)
print("Degree of Freedom:-",ddof)
alpha = 0.05
```

```python

from scipy.stats import chi2
chi_square=sum([(o-e)**2./e for o,e in zip(Observed_Values,Expected_Values)])
chi_square_statistic=chi_square[0]+chi_square[1]
```

```python
print("chi-square statistic:-",chi_square_statistic)
```

```python
critical_value=chi2.ppf(q=1-alpha,df=ddof)
print('critical_value:',critical_value)
```

```python
#p-value
p_value=1-chi2.cdf(x=chi_square_statistic,df=ddof)
print('p-value:',p_value)
print('Significance level: ',alpha)
print('Degree of Freedom: ',ddof)
print('p-value:',p_value)
```

```python
if chi_square_statistic>=critical_value:
    print("Reject H0,There is a relationship between 2 categorical variables")
else:
    print("Retain H0,There is no relationship between 2 categorical variables")
    
if p_value<=alpha:
    print("Reject H0,There is a relationship between 2 categorical variables")
else:
    print("Retain H0,There is no relationship between 2 categorical variables")
```

Above we are rejecting null hypothesis means their is relationship between Gender and SMOKE variables 
and the person who smokes is more likely to be a man

# Anova Test(F-Test)

It is a statistical test that is used to compare the variances of two or more groups or populations. It is commonly employed
in the field of analysis of variance (ANOVA) to determine if there are significant differences between the means of multiple groups.

We wanted to test whether Weight differs based on some categorical variable like CALC, we
have to compare the means of each level or group the variable
Here we have  4 different category for people consuming alcohol (i.e •No •Sometimes •Frequently •Always ) 
and need to check whether all 4 group are similar or no

```python
df_anova = dataset[['Weight','CALC']]
```

```python
grps = pd.unique(df_anova.CALC.values)
grps
```

```python
d_data = {grp:df_anova['Weight'][df_anova.CALC == grp] for grp in grps}
d_data
```

```python
F, p = stats.f_oneway(d_data['no'], d_data['Sometimes'], d_data['Frequently'],d_data['Always'])
```

```python
p
```

```python
if p<0.05:
    print("reject null hypothesis")
else:
    print("accept null hypothesis")
```

Rejecting null hypothesis means that all 4 group are not similar(in terms of weight) and weight varies according
to the drinking habit.
